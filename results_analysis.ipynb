```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
import plotly.graph_objects as go
import plotly.figure_factory as ff
```

```python
np.version.version #check which version of NumPy is used
```

```python
with open('Result.txt') as f:
    lines = f.readlines() #returns a list containing each line in the file as a list item
```

```python
#definition of function: convert string inside list to int
def ev(l: list[str]) -> list:
    return [int(el) for el in l]
    
X = np.array([ev(l.split()) for l in lines]) #split(): splits a string into a list
```

```python
X.shape #check shape of the array X
```

```python
C = [] #stores localization of  particles
for j in range(X.shape[0]):
    l = X[j] # line corresponds to system's state
    C += np.where(l==1)[0].tolist() #converts series to list
```

```python
fig = go.Figure(data=[go.Histogram(x=C)])
fig.update_layout(title="Index distribution of particles",
                  xaxis_title="Index",
                  yaxis_title="Count")
fig.show()
```

```python
fig = ff.create_distplot([C], group_labels=['Index'])
fig.update_layout(title="Index distribution of particles")
fig.show()
```
